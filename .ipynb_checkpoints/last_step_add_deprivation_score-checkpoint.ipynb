```python
# Add deprivation score from 2021 onto the residential matched addresses
## INPUT: residential matched addresses, Household deprivation LSOA 2021.csv, PCD_OA21_LSOA21_MSOA21_LAD_NOV23_UK_LU 2.csv
## OUTPUT: CSV with deprivation score columns 

import pandas as pd
```

```python
# take the matched data and add the deprivation score... 
address_file = "/Users/christine/Documents/_UCL_grad school/research/research_pd_24/Batch3c_address_matched_2006.csv"
addresses = pd.read_csv(address_file)
```

```python
dep_file = "/Users/christine/Documents/_UCL_grad school/research/research_pd_24/data/Household deprivation LSOA 2021.csv"
dep_data = pd.read_csv(dep_file, skiprows =6 )
```

```python
#need to load in the conversion from LSOA to Postcode
file = "/Users/christine/Documents/_UCL_grad school/research/research_pd_24/data/PCD_OA21_LSOA21_MSOA21_LAD_NOV23_UK_LU 2.csv"
df = pd.read_csv(file, encoding='latin-1', low_memory=False)
```

```python
#drop the rows where lsoa21cd is null... 
df= df.dropna(subset=['lsoa21cd', 'pcds'])
```

```python
#drop columns we do not need
df = df.drop(columns=['pcd7','pcd8','dointr','doterm','usertype','oa21cd','msoa21cd','ladcd','lsoa21nm','msoa21nm','ladnm','ladnmw'])
```

```python
# bring the dep data _ merge with the df but only keep the pcds column 
# merge on 'lsoa21cd'  = 'mnemonic'

#merge the dep data with the LSOA data to get the post codes
dep_data = dep_data.merge(df, how = 'left', left_on='mnemonic', right_on='lsoa21cd')

#keep the LSOA and postcodes 
dep_data_to_merge = dep_data.drop(columns= ['2021 super output area - lower layer','mnemonic','lsoa21cd'])
```

```python
# Lastly, take the addresses and merge on postcode with the deprivation data
addresses = addresses.merge(dep_data_to_merge, how = 'left', left_on = 'postcode', right_on = 'pcds') 
```
